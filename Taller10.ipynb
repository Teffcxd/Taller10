```python
def contar(n, vec):
  count=0
  for i in vec:
    if i==n:
      count+=1
  return count
```

```python
print('Los intentos de suicidio en el barrio "Centro" son: ',contar('CENTRO', lista_barrio))
print('Los intentos de suicidio en el barrio "Los muiscas" son : ',contar('LOS MUISCAS', lista_barrio))
print('Los intentos de suicidio en el barrio "Fuente" son: ',contar('FUENTE', lista_barrio))
print('Los intentos de suicidio en el barrio "Antonia Santos" son: ',contar('ANTONIA SANTOS', lista_barrio))
print('Los intentos de suicidio en el barrio "Asis Boyacense" son: ',contar('ASIS BOYACENSE', lista_barrio))
print('Los intentos de suicidio en el barrio "La Maria" son: ',contar('LA MARIA', lista_barrio))
print('Los intentos de suicidio en el barrio "20 de Julio"  es: ',contar('20 JULIO', lista_barrio))
print('Los intentos de suicidio en el barrio "La florida" son: ',contar('LA FLORIDA', lista_barrio))

aux=[contar('CENTRO', lista_barrio),contar('LOS MUISCAS', lista_barrio),
     contar('FUENTE', lista_barrio),contar('ANTONIA SANTOS', lista_barrio), contar('ASIS BOYACENSE', lista_barrio),
     contar('LA MARIA', lista_barrio) , contar('20 JULIO', lista_barrio),contar('LA FLORIDA', lista_barrio)]
```

```python
num_barrio=[]
for i in range(len(lista_barrio)):
  num_barrio.append(contar(lista_barrio[i],lista_barrio))

print(num_barrio)
print(len(num_barrio))
```

```python
barrio_de_residencia=['Centro','Muiscas','Fuente', 'Antonia', "ASIS", "MARIA","20 JULIO","FLORIDA"]
plt.bar(barrio_de_residencia, aux ,width=0.8, color='crimson')
plt.xlabel('Barrios')
plt.ylabel('intentos')
plt.title('Análisis sobre la frecuencia de intentos de suicidios en los barrios')
plt.grid()
plt.show()
```

En el estudio de los 8 barrios encontramos que existen dos potenciales barrios en los cuales los indices alcanzan a superar los 25 casos de intentos de suicidios
